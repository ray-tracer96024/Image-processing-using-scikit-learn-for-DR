```python
# Show image in RGB
store_image = skimage.io.imshow('3_left.jpeg')
```

```python
# Show shape of the image
store_image_array = skimage.io.imread('3_left.jpeg')
store_image_array.shape
```

```python
# Show image in grayscale
grayscale_image = skimage.color.rgb2gray(store_image_array)
skimage.io.imshow(grayscale_image)
```

```python
image_rescaled = skimage.transform.rescale(grayscale_image, 1.0/2.0, anti_aliasing = False)
image_resized = skimage.transform.resize(grayscale_image, (grayscale_image.shape[0] / 2, grayscale_image.shape[1] / 2), anti_aliasing = True)
image_downscaled = skimage.transform.downscale_local_mean(grayscale_image, (5, 4))
```

```python
import matplotlib.pyplot as plt
figure, axes = plt.subplots(nrows=2, ncols=2)
ax = axes.ravel()

ax[0].imshow(grayscale_image, cmap = 'gray')
ax[0].set_title('Original')

ax[1].imshow(image_rescaled, cmap='gray')
ax[1].set_title("Rescaled image (aliasing)")
print(image_rescaled.shape)

ax[2].imshow(image_resized, cmap='gray')
ax[2].set_title("Resized image (no aliasing)")
print(image_resized.shape)

ax[3].imshow(image_downscaled, cmap='gray')
ax[3].set_title("Downscaled image (no aliasing)")
print(image_downscaled.shape)

ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```

```python
image_new_size = skimage.transform.resize(grayscale_image, (1080, 1920), anti_aliasing = False)
skimage.io.imshow(image_new_size)
skimage.io.imsave(fname = 'D:/MP_1_2/MinorProject_material/Project_implementation/resized.jpeg', arr = skimage.img_as_ubyte(image_new_size))
```

```python
filtered_grayscale_image = skimage.filters.gaussian(image_new_size, sigma = 1)
skimage.io.imshow(filtered_grayscale_image)
skimage.io.imsave(fname = 'D:/MP_1_2/MinorProject_material/Project_implementation/filtered.jpeg', arr = skimage.img_as_ubyte(filtered_grayscale_image))
```

```python
canny_grayscaled_image = skimage.feature.canny(filtered_grayscale_image, sigma = 0.1)
skimage.io.imshow(canny_grayscaled_image)
skimage.io.imsave(fname = 'D:/MP_1_2/MinorProject_material/Project_implementation/canny_img.jpeg', arr = skimage.img_as_ubyte(canny_grayscaled_image))
```
